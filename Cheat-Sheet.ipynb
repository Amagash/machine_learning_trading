## Lesson 2: Reading and plotting stock data

```python
import pandas as pd

df = pd.read_csv("bitflyerJPY_1-min_data_2017-07-04_to_2018-03-27.csv")
```

```python
print(df.head())
```

```python
print(df[10:21])
```

```python
print(df.tail())
```

```python
def get_max_close(df):
    print("Max Close")
    return df['Close'].max()
    
get_max_close(df)
```

```python
def get_mean_volume(df):
    print("Mean Volume")
    return df['Volume_(BTC)'].mean()

get_mean_volume(df)
```

### Plotting stock price data

```python
import matplotlib.pyplot as plt

def plot(df):
    df['Close'].plot()
    plt.show()

plot(df)
```

### Plot two columns

```python
def plot_two_col(df):
    df[['Low', 'High']].plot()
    plt.show()

plot_two_col(df)
```

## Working with multiple stocks

```python
import datetime

def convert_timestamp_to_date(timestamp):
    date = datetime.datetime.fromtimestamp(int(timestamp)).strftime('%Y-%m-%d %H:%M:%S')
    return date
```

```python
def date_range_function(df):
    start_date = convert_timestamp_to_date('1499155860')
    end_date = convert_timestamp_to_date('1499156460')
    print(start_date, end_date)
    dates = pd.date_range(start_date, end_date)
    return dates
date_range_function(df)
```
